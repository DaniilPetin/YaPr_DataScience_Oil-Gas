```python
def duplicates_search(df, df_name):
    
    print()
    print(f'\033[1mЧисло неявных дубликатов в датасете {df_name} равно {len(df[df.duplicated(["id"])])}')    
    print()
    print(f'\033[1mНеявные дубликатов в датасете {df_name} по столбцу id:')
    
    duplicated_data = df[df.duplicated(['id'])]
    duplicated_id = duplicated_data['id'].unique()
    display(df.loc[df['id'].isin(duplicated_id)])
```

В датафрейме `geo_data_0` 10 невяных дубликатов. Исправим это, добавив повторяющимся `id` суффиксы _1, _2, _3 и т.д.

```python
def add_suffix_to_duplicates(df: pd.DataFrame, id_column: str = 'id'):
        
    df['_duplicate_count'] = df.groupby(id_column).cumcount() # столбец для подсчета дубликатов
    
    mask = df['_duplicate_count'] > 0 # маска для дубликатов
    
    original_ids = df.loc[mask, id_column].copy() # оригинальные id для вывода
    
    # добавление суффиксов к дубликатам
    df.loc[mask, id_column] = (
        df.loc[mask, id_column] + '_' + df.loc[mask, '_duplicate_count'].astype(str)
    )
    
    df = df.drop(columns=['_duplicate_count'])
    
    changed_rows = df[mask].copy()
    changed_rows['original_id'] = original_ids
    print("\nБыли изменены следующие строки:")
    print(changed_rows[[id_column, 'original_id'] + 
                        [col for col in df.columns if col != id_column]].to_string(index=True))
    print(f"\nВсего изменено id: {len(changed_rows)}")
    
    return df
```

Как можно увидеть из ящика с усами, в признаке `f2` таблицы `geo_data_0` присутствуют выбивающиеся из общего распределения значения. Для того, чтобы принять решения, удалить их или оставить в данных, нужно выяснить, сколько их относительно общего объема данных

```python
def delete_anomalies(df, column, df_name):
    
    # квантили:
    q25 = df[column].quantile(0.25)
    q75 = df[column].quantile(0.75)
    
    # границы допустимой области без выбросов:
    left_border = q25 - 1.5 * (q75-q25)
    right_border = q75 + 1.5  * (q75-q25)
    
    outside_quantiles = df[(df[column] < left_border) | (df[column] > right_border)]
    
    mask = (df[column] >= left_border) & (df[column] <= right_border) # маска сохраняемых строк
    
    percentage = (len(outside_quantiles) / len(df)) * 100
    
    df = df.drop(df[~mask].index, inplace=True)
    
    print(f'Процент выбросов таблицы {df_name} относительно столбца {column} равен {percentage:.3f}%')
    print()
    print(f'Границы: [{left_border:.2f}, {right_border:.2f}]')
    print()
    print('Выбивающиеся данные были удалены')
    
    return df
```

```python
delete_anomalies(geo_data_0, 'f2', 'geo_data_0')
```

## 5. Расчёт прибыли и рисков 

### Расчет прибыли

```python
def profit(values, predictions):
    predictions_sorted = predictions.sort_values(ascending=False)
    values_best = values[predictions_sorted.index][:pts_best]
    values_sum = values_best.sum()
    profit = one_unit_profit * values_sum - budget
    return profit
```

```python
print('Прибыль для', pts_best, 'лучших скважин 1-го региона равна', round(profit(data_0_predicted['validation'], data_0_predicted['prediction'])), 'руб.')
print('Прибыль для', pts_best, 'лучших скважин 2-го региона равна', round(profit(data_1_predicted['validation'], data_1_predicted['prediction'])), 'руб.')
print('Прибыль для', pts_best, 'лучших скважин 3-го региона равна', round(profit(data_2_predicted['validation'], data_2_predicted['prediction'])), 'руб.')
```

### Расчет рисков

Для того, чтобы найти распределение прибыли в каждом регионе и оценить риск убытков, воспользуемся технологией *Bootstrap*.

```python
def bootstrap (values, predictions):
    state = np.random.RandomState(12345)

    profit_values = []
    bootstrap_samples = 1000
    
    for i in range(bootstrap_samples):
        values_subsample = values.sample(n=500, replace=True, random_state=state)
        predictions_subsample = predictions[values_subsample.index]
 
        profit_values.append(profit(values_subsample, predictions_subsample))

    profit_values = pd.Series(profit_values)
    
    return profit_values
```

```python
profits_0 = bootstrap(data_0_predicted['validation'], data_0_predicted['prediction'])
profits_1 = bootstrap(data_1_predicted['validation'], data_1_predicted['prediction'])
profits_2 = bootstrap(data_2_predicted['validation'], data_2_predicted['prediction'])
```
